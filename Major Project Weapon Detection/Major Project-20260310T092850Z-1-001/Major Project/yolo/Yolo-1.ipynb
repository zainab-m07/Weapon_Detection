```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_bounding_box(image_path, bbox):
    # Read the image
    image = cv2.imread(image_path)

    # Convert BGR to RGB (OpenCV uses BGR by default)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Create a figure and axis
    fig, ax = plt.subplots(1)

    # Display the image
    ax.imshow(image_rgb)

    # Create a rectangle patch
    rect = patches.Rectangle(
        (bbox[0], bbox[1]),  # (x, y) coordinates of the top-left corner of the rectangle
        bbox[2] - bbox[0],    # Width of the rectangle
        bbox[3] - bbox[1],    # Height of the rectangle
        linewidth=2,
        edgecolor='r',        # Red color for the bounding box
        facecolor='none'      # No fill color
    )

    # Add the rectangle patch to the axis
    ax.add_patch(rect)

    # Display the image with the bounding box
    plt.show()

# Example usage
image_path = '/content/drive/MyDrive/Major Project/yolo/Dataset/Train/images/0045a7e07dd4fa0e.jpg'
bounding_box = [5.6017e+02, 3.3523e+02, 7.0484e+02, 3.9538e+02]  # Format: [x_min, y_min, x_max, y_max]


plot_bounding_box(image_path, bounding_box)
```
